# Classifying PPG Dataset with Machine Learning

Dataset Reference: https://ieeexplore.ieee.org/document/8529266

```python
import numpy as np
import pandas as pd
import glob
import scipy.io
import matplotlib.pyplot as plt
```

```python
# Retrieve all the datasets
file_list = glob.glob("Data/gyro_acc_ppg-master/*.mat")
mat = scipy.io.loadmat(file_list[0])

print(f"Size of ECG Numpy Array: {np.size(mat['bpmECG'])}")
subject_1_ecg = pd.DataFrame(mat['timeECG'], columns=['time'])
subject_1_ecg = pd.concat([subject_1_ecg, pd.DataFrame(mat['bpmECG'], columns=['ecg_ch0'])], axis=1)
print(f"Size of ECG Dataframe: {subject_1_ecg.size}")

(subject_1_ecg
 .plot(kind='scatter', x='time', y='ecg_ch0'))
plt.xlabel('Time (s)')
plt.ylabel('ECG (bpm)')
plt.title('ECG - Ground Truth')
```

```python
acc_fs = 50 # hz
print(f"Size of ACC Numpy Array: {np.size(mat['sigAcc'])}")
subject_1_acc = pd.DataFrame(mat['sigAcc'].T, columns=['acc_ch0', 'acc_ch1', 'acc_ch2'])
subject_1_acc = pd.concat([pd.DataFrame({'time': np.linspace(0, len(subject_1_acc) / acc_fs,  len(subject_1_acc))}), subject_1_acc], axis=1)
print(f"Size of ACC Dataframe: {subject_1_acc.size}")

(subject_1_acc
 .plot(kind='line',alpha=0.85))
plt.xlabel('Sample')
plt.ylabel('ACC')
plt.title('ACC - 3 Channel')
```

```python
print(f"Size of PPG Numpy Array: {np.size(mat['sigPPG'])}")
subject_1_ppg = pd.DataFrame(mat['sigPPG'].T, columns=['ppg_ch0', 'ppg_ch1', 'ppg_ch2'])
print(f"Size of PPG Dataframe: {subject_1_ppg.size}")

(subject_1_ppg
 .plot(kind='line', alpha=0.85))
plt.xlabel('Sample')
plt.ylabel('PPG (bpm)')
plt.title('PPG - 3 Channel')
```
